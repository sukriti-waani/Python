## Flask

Flask is a lightweight Python library used to create web applications like websites and APIs.

“Lightweight” means it is simple, easy to learn, and gives only what is needed.

Why Flask is used:
- To build websites
- To create REST APIs
- To connect Python code with the web

## Flask Framework
The Flask framework is the complete set of tools that Flask provides to build web apps.

It includes:
- URL routing
- Request & response handling
- Template support (Jinja2)
- Error handling
- Session & cookies

##### Features of Flask Framework:
- Lightweight:	Less complexity
- Flexible:	Add only what you want
- Modular:	Use extensions easily
- Beginner-friendly:	Easy syntax

- Flask Request Flow
. Browser → Flask → Python Function → Response → Browser

## WSGI (Web Server Gateway Interface)
WSGI is a standard rule that allows web servers to communicate with Python web apps.

##### Why WSGI is needed?
- Browser cannot talk to Python directly
- Server understands HTTP
- Python app understands Python

#### Flow Using WSGI
- Browser
-    ↓
- Web Server (Gunicorn / uWSGI)
-    ↓
- WSGI
-    ↓
- Flask App
-    ↓
- Response back to Browser

## Jinja2
Jinja2 is a template engine used in Flask to create dynamic HTML pages.

It mixes:
- HTML + Python data

##### Why Jinja2 is used?
- To avoid hard-coding HTML
- To show dynamic data (username, tables, lists)

Feature          	               Use
