## 1️⃣ Introduction to Probability
**Probability** is the measure of how likely an event is to occur.

Probability values always lie between **0 and 1**:

- **P = 0** → Event will not occur  
- **P = 1** → Event will surely occur  


### Basic Probability Formula
P(Event) = Number of favorable outcomes / Total number of possible outcomes


## Sample Space

The **sample space** is the set of all possible outcomes of an experiment.

### Example: Tossing a Coin
Sample Space = {H, T}


### Probability Examples

### Example 1: Tossing a Coin
P(H) = 1/2
P(T) = 1/2

### Example 2: Rolling a Dice
Sample Space = {1, 2, 3, 4, 5, 6}

P(X = 1) = 1/6


## Events in Probability

An **event** is a subset of the sample space.

### Examples:
- Getting **Head** in a coin toss  
- Getting **5** in a dice roll  


## Mutually Exclusive Events
Two events are **mutually exclusive** if **they cannot occur at the same time**.

### Example
Tossing a coin:
- Event A = Head (H)  
- Event B = Tail (T)  

Both cannot happen together.


## Addition Rule for Mutually Exclusive Events

### Example: Coin Toss
P(H or T) = P(H) + P(T)
= 1/2 + 1/2
= 1

### Example: Rolling a Dice
Event A = Getting 1  
Event B = Getting 5  
P(1 or 5) = P(1) + P(5)
= 1/6 + 1/6
= 2/6
= 1/3


## Non–Mutually Exclusive Events

### Definition
Two events are **non-mutually exclusive** if **they can occur at the same time**.  
They have **common outcomes**.

### Example
Drawing a card:
- Event A = King (K)  
- Event B = Heart (♥)  

The **King of Hearts** belongs to both events.

## Addition Rule for Non-Mutually Exclusive Events
P(A or B) = P(A) + P(B) − P(A and B)

### Example: Cards from a Deck
Total cards = 52  

P(K) = 4/52
P(♥) = 13/52
P(K and ♥) = 1/52


Now apply formula:
P(K or ♥) = 4/52 + 13/52 − 1/52
= 16/52


#### Why Do We Subtract P(A and B)?
In **non-mutually exclusive** events, the common outcomes get counted **twice**.  
Subtracting **P(A and B)** removes the double-counting.


#### Important Probability Formulas (Summary)

P(A) = Favorable outcomes / Total outcomes

P(A or B) [Mutually Exclusive] = P(A) + P(B)

P(A or B) [Non-Mutually Exclusive]
= P(A) + P(B) − P(A and B)


#### Summary

- Probability measures likelihood of an event  
- Sample space = all possible outcomes  
- Mutually exclusive events cannot occur together  
- Non-mutually exclusive events can occur together  
- Addition rule depends on the type of events
