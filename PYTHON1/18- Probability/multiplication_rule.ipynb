## What is Multiplication Rule?
The **Multiplication Rule** is used to find the probability of **two events occurring together**.

It is written as:
P(A and B)

The rule depends on whether the events are:
- **Independent**
- **Dependent**

##  Independent Events
Two events are said to be **independent** if the occurrence of one event **does not affect** the occurrence of the other event.

#### Examples of Independent Events
- Tossing a coin twice  
- Tossing a coin and rolling a dice  


###  Multiplication Rule for Independent Events
P(A and B) = P(A) × P(B)


#### Example 1: Tossing a Coin
Event A = Getting Head (H)  
Event B = Getting Tail (T)  

P(H) = 1/2
P(T) = 1/2

Applying multiplication rule:
P(H and T) = P(H) × P(T)
= 1/2 × 1/2
= 1/4

##  Dependent Events
Two events are **dependent** if the occurrence of one event **affects the probability** of the other event.

In dependent events, we use **conditional probability**.


#### Example of Dependent Events
Taking cards from a deck **without replacement**:

- Event A = Drawing a King (K)
- Event B = Drawing a Queen (Q) after the King is removed


###  Conditional Probability
Conditional probability means **probability of B given that A has already occurred**.

It is written as:
P(B | A)

###  Multiplication Rule for Dependent Events
P(A and B) = P(A) × P(B | A)


### Example: Cards from a Deck

Total cards = 52  

Event A = King  
Event B = Queen after King is removed  

P(K) = 4/52
P(Q | K) = 4/51

Applying the rule:
P(K and Q) = P(K) × P(Q | K)
= 4/52 × 4/51


##  When to Use Which Rule?

-  Use **independent multiplication rule** when events do **not affect** each other  
-  Use **dependent multiplication rule** when probability **changes** after first event  

## Important Multiplication Formulas (Summary)

P(A and B) = P(A) × P(B) → Independent Events

P(A and B) = P(A) × P(B | A) → Dependent Events


##  Summary

- Multiplication rule finds probability of two events together  
- Independent events do not affect each other  
- Dependent events affect future probabilities  
- Conditional probability is used in dependent events  
- Correct formula depends on the type of events
