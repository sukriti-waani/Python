### 1. Program
A program is a set of instructions written in a file, stored on disk.

Example:
A Python file calculator.py is a program.

### 2. Process
A process is a program in execution.
When we run a program, the operating system creates a process.

Example:
When we run:   python calculator.py .
A process is created in RAM to execute that program.

### 3. Thread
A thread is a smallest unit of execution inside a process.
A process can have multiple threads running tasks simultaneously.

Example:
A web browser process may have threads like:
one thread loading a webpage
another thread playing a video
another handling user input
