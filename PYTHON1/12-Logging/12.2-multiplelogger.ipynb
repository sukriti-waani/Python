### Logging With Multiple Loggers

We can create multiple loggers for different parts of our application.

```python
import logging                      # Import Python's logging module

# Create a logger for module1
logger1 = logging.getLogger("module1")   # Create a logger named "module1"
logger1.setLevel(logging.DEBUG)          # This logger will log DEBUG and above messages

# Create a logger for module2
logger2 = logging.getLogger("module2")   # Create a logger named "module2"
logger2.setLevel(logging.WARNING)        # This logger will log WARNING and above messages

# Configure logging settings
logging.basicConfig(
  level = logging.DEBUG,                 # Minimum level to log (shows DEBUG and above)
  format = '%(asctime)s - %(name)s - %(levelname)s - %(message)s',  # Log message format
  datefmt = '%Y-%m-%d %H:%M:%S'          # Format for date and time
)
```

```python
# log messages with different loggers

logger1.debug("This is a debug message from module1")
# Uses logger1 → allowed because logger1 level = DEBUG
# DEBUG message will be printed

logger1.info("This is an info message from module1")
# Uses logger1 → INFO is above DEBUG
# So this message will also be printed

logger2.warning("This is a warning message from module2")
# Uses logger2 → logger2 level = WARNING
# WARNING is allowed, so this will be printed

logger2.error("This is an error message from module2")
# Uses logger2 → ERROR is above WARNING
# So this message will also be printed
```
