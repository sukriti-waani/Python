In statistics, a **hypothesis** is a **statement or assumption** made about a **population parameter** (such as mean or proportion).

Since we usually cannot study the entire population, we:
- Take a **sample**
- Perform **statistical analysis**
- Draw conclusions about the **population**

This process belongs to **Inferential Statistics**.

## 2Ô∏è‚É£ Inferential Statistics

**Inferential statistics** is the branch of statistics that helps us:
- Make **conclusions**
- Draw **inferences**
- Take **decisions**

about a **population**, based on **sample data**.

From the notes:
- Sample ‚Üí Statistical Analysis ‚Üí Conclusion ‚Üí Population

This process is called **Hypothesis Testing**.


## 3Ô∏è‚É£ What is Hypothesis Testing?
**Hypothesis testing** is a **statistical mechanism** used to:
- Test an assumption about a population
- Decide whether to **accept or reject** that assumption based on data

It helps answer questions like:
- Is a person guilty or not?
- Is a coin fair or biased?
- Is the average different from what is claimed?

## 5Ô∏è‚É£ Basic Structure of Hypothesis Testing

Hypothesis testing always involves:
1. Making an initial assumption  
2. Collecting sample data  
3. Performing statistical analysis  
4. Taking a decision  

The decision is based on **probability and evidence**, not guesswork.

# üìò Null Hypothesis (H‚ÇÄ)

## 1Ô∏è‚É£ Definition of Null Hypothesis

The **Null Hypothesis**, denoted by **H‚ÇÄ**, is a **statement of no effect, no difference, or no change** in the population.

It represents the **default or initial assumption** that nothing unusual has happened.


## 2Ô∏è‚É£ Meaning of Null Hypothesis

The null hypothesis assumes that:
- Any observed difference in sample data is due to **random chance**
- The population parameter has **not changed**
- There is **no significant effect**

In hypothesis testing, **H‚ÇÄ is assumed to be true initially**.


## 3Ô∏è‚É£ Why Do We Start with Null Hypothesis?

We start with H‚ÇÄ because:
- It is a **conservative assumption**
- It avoids jumping to conclusions
- It requires **strong evidence** to be rejected

This is similar to the legal principle:
- A person is assumed **innocent until proven guilty**
