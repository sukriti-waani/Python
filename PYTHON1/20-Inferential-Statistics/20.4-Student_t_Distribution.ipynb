## 6Ô∏è‚É£ Nature of Student‚Äôs t-Distribution

- It is **symmetric** like the normal distribution
- It is **bell-shaped**
- It has **heavier tails** than the normal distribution
- As sample size increases, it **approaches normal distribution**

This happens because uncertainty decreases with more data.

For Student‚Äôs t-distribution:
