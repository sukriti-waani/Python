## 1Ô∏è‚É£ Background: Reality vs Decision

In hypothesis testing, there are **two things** to consider:

### Reality
- Null hypothesis (H‚ÇÄ) is **true**
- Null hypothesis (H‚ÇÄ) is **false**

### Decision
- We **reject** the null hypothesis
- We **retain (fail to reject)** the null hypothesis

Because of these two possibilities, **four outcomes** are possible.

## 2Ô∏è‚É£ Outcome 1: Correct Decision (Good)

### Situation
Reality: Null hypothesis is false
Decision: We reject the null hypothesis

### Explanation
- The null hypothesis is actually false
- We correctly reject it
- This is a **good decision**

This means our statistical test worked correctly.

## 3Ô∏è‚É£ Outcome 2: Type I Error

### Situation
Reality: Null hypothesis is true
Decision: We reject the null hypothesis

### Explanation
- The null hypothesis is actually true
- But we reject it incorrectly
- This is called a **Type I Error**

### Definition
A **Type I Error** occurs when we **reject a true null hypothesis**.

## 4Ô∏è‚É£ Outcome 3: Type II Error

### Situation
Reality: Null hypothesis is false
Decision: We retain the null hypothesis

### Explanation
- The null hypothesis is actually false
- But we fail to reject it
- This is called a **Type II Error**

### Definition
A **Type II Error** occurs when we **fail to reject a false null hypothesis**.

## 5Ô∏è‚É£ Outcome 4: Correct Decision (Good)

### Situation
Reality: Null hypothesis is true
Decision: We retain the null hypothesis

### Explanation
- The null hypothesis is true
- We correctly retain it
- This is a **good decision**
