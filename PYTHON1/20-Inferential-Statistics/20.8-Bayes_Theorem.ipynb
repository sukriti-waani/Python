## 1Ô∏è‚É£ Introduction to Bayesian Statistics

**Bayesian statistics** is an approach to **data analysis and parameter estimation** based on **Bayes‚Äô theorem**.

It uses:
- Prior knowledge
- Observed data
- Probability theory

to update our belief about an event.

## 2Ô∏è‚É£ Probability and Types of Events

In probability theory, events can be classified into:
- **Independent events**
- **Dependent events**

This classification is important to understand Bayes‚Äô theorem.

## 3Ô∏è‚É£ Independent Events
Two events are called **independent** if the occurrence of one event **does not affect** the occurrence of the other event.

### Example 1: Rolling a Dice

Sample space:
{1, 2, 3, 4, 5, 6}

Probability of each outcome:
P(1) = 1/6
P(2) = 1/6
...
P(6) = 1/6

Each outcome is independent of the others.

### Example 2: Tossing a Coin

P(H) = 0.5, P(T) = 0.5

Getting head does not affect the probability of getting tail.

## 4Ô∏è‚É£ Dependent Events

### Definition
Two events are called **dependent** if the occurrence of one event **affects** the probability of the other event.

### Example (from notes)

A box contains colored balls.

Let:
- Event R = picking a **Red** ball  
- Event Y = picking a **Yellow** ball  

From the notes:
P(R) = 2/5
P(Y) = 3/4 (after red is chosen)

This shows dependency.

## 5Ô∏è‚É£ Conditional Probability
**Conditional probability** is the probability of an event occurring **given that another event has already occurred**.

It is written as:
P(A | B)

Meaning:
- Probability of A given B is true

### Formula for Conditional Probability

P(A | B) = P(A and B) / P(B)

Similarly:
P(B | A) = P(A and B) / P(A)


## 6Ô∏è‚É£ Joint Probability Property

From probability rules:
P(A and B) = P(B and A)

Using conditional probability:
P(A and B) = P(A) ¬∑ P(B | A)
P(A and B) = P(B) ¬∑ P(A | B)

## 7Ô∏è‚É£ Derivation of Bayes‚Äô Theorem

Since:
P(A) ¬∑ P(B | A) = P(B) ¬∑ P(A | B)

Rearranging the equation:
P(B | A) = [P(B) ¬∑ P(A | B)] / P(A)

This formula is called **Bayes‚Äô Theorem**.

## 8Ô∏è‚É£ Bayes‚Äô Theorem Formula

P(B | A) = (P(B) ¬∑ P(A | B)) / P(A)

Where:
- A, B are events
- P(B | A) ‚Üí Posterior probability
- P(B) ‚Üí Prior probability
- P(A | B) ‚Üí Likelihood
- P(A) ‚Üí Evidence

## 9Ô∏è‚É£ Meaning of Terms (from notes)

- `P(A | B)`  
  Probability of event **A given B is true**

- `P(B | A)`  
  Probability of event **B given A is true**

- `P(A)` and `P(B)`  
  Independent probabilities of events A and B

### Independent Variables (Features)

### Dependent Variable (Output)

## 1Ô∏è‚É£1Ô∏è‚É£ Bayes‚Äô Theorem for Dataset

The probability of price given features is:

Using Bayes‚Äô theorem:

P(y | x‚ÇÅ, x‚ÇÇ, x‚ÇÉ) = [P(y) ¬∑ P(x‚ÇÅ, x‚ÇÇ, x‚ÇÉ | y)] / P(x‚ÇÅ, x‚ÇÇ, x‚ÇÉ)

This equation is highlighted in the notes.

## 1Ô∏è‚É£2Ô∏è‚É£ Interpretation in Machine Learning

- `P(y)` ‚Üí Prior belief about price  
- `P(x‚ÇÅ, x‚ÇÇ, x‚ÇÉ | y)` ‚Üí Likelihood of features given price  
- `P(y | x‚ÇÅ, x‚ÇÇ, x‚ÇÉ)` ‚Üí Updated belief (prediction)

This forms the foundation of **Bayesian models** such as:
- Naive Bayes classifier
- Bayesian inference
