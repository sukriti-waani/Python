## 1Ô∏è‚É£ Introduction to ANOVA
**ANOVA (Analysis of Variance)** is a statistical test used to determine whether there are **significant differences between the means of three or more groups**.

Instead of comparing means directly, ANOVA compares **variations (variances)** to make decisions.

## 2Ô∏è‚É£ Why ANOVA is Used

If we compare multiple group means using repeated t-tests:
- Probability of error increases
- Results become unreliable

ANOVA solves this by:
- Performing **one single test**
- Controlling the **Type I error**

## 3Ô∏è‚É£ Hypotheses in ANOVA
### Null Hypothesis (H‚ÇÄ)
All group means are equal
Œº‚ÇÅ = Œº‚ÇÇ = Œº‚ÇÉ = ... = Œº‚Çñ

### Alternative Hypothesis (H‚ÇÅ)
At least one group mean is different

## 4Ô∏è‚É£ Basic Idea Behind ANOVA
ANOVA compares:
- **Variation between groups**
- **Variation within groups**

If the variation **between groups** is much larger than the variation **within groups**, then the group means are considered significantly different.

## 5Ô∏è‚É£ Test Statistic in ANOVA (F-Test)

ANOVA uses the **F-statistic**, defined as:

F = Variance Between Groups / Variance Within Groups

If F is large ‚Üí strong evidence against H‚ÇÄ.

## 6Ô∏è‚É£ Assumptions of ANOVA
ANOVA is valid only when the following assumptions are satisfied:

### 1. Independence
- Observations are independent of each other

### 2. Normality
- Data in each group is approximately normally distributed

### 3. Homogeneity of Variance
- All groups have **equal variances**

This is also called **homoscedasticity**.

## 7Ô∏è‚É£ Types of ANOVA

### 1Ô∏è‚É£ One-Way ANOVA
- Used when there is **one independent factor**
- Compares means of **three or more groups**

Example:
Effect of teaching method on student performance


### 2Ô∏è‚É£ Two-Way ANOVA
- Used when there are **two independent factors**
- Studies:
  - Main effect of factor 1
  - Main effect of factor 2
  - Interaction effect

Example:
Effect of teaching method and gender on marks

### 3Ô∏è‚É£ Repeated Measures ANOVA

- Same subjects measured multiple times
- Used in **time-based or condition-based studies**

Example:
Blood pressure before, during, and after treatment

### 4Ô∏è‚É£ MANOVA (Multivariate ANOVA)

- Used when there are **multiple dependent variables**

Example:
Effect of diet on weight and cholesterol level

Total Variation = Between-Group Variation + Within-Group Variation
