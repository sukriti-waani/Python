## 1Ô∏è‚É£ Definition of Alternative Hypothesis

The **Alternative Hypothesis**, denoted by **H‚ÇÅ** or **H‚Çê**, is a **statement that contradicts the null hypothesis**.

It represents the claim that:
- There **is an effect**
- There **is a difference**
- The assumption made in the null hypothesis is **not true**

## 2Ô∏è‚É£ Meaning of Alternative Hypothesis
The alternative hypothesis suggests that:
- The observed result is **not due to random chance**
- There is a **real change or effect** in the population
- The population parameter is **different from what H‚ÇÄ claims**

It reflects what we want to **prove using data**.

## 3Ô∏è‚É£ Relationship Between H‚ÇÄ and H‚ÇÅ

- **H‚ÇÄ** and **H‚ÇÅ** are always **opposites**
- If H‚ÇÄ is rejected, H‚ÇÅ is accepted
- If H‚ÇÄ is not rejected, H‚ÇÅ cannot be proven

From the notes:
H‚ÇÄ : Coin is fair
H‚ÇÅ : Coin is not fair

### Example 1: Coin Toss
H‚ÇÄ : Coin is fair
H‚ÇÅ : Coin is not fair

This means:
- Probability of Head is **not equal to 0.5**
- Probability of Tail is **not equal to 0.5**

## 6Ô∏è‚É£ Key Points about Alternative Hypothesis
- Denoted by **H‚ÇÅ or H‚Çê**

# üìò P-Value

## 1Ô∏è‚É£ Definition of P-Value

The **p-value** is the **probability of obtaining a result as extreme as (or more extreme than) the observed result**, assuming that the **null hypothesis (H‚ÇÄ) is true**.

In simple words:
- It tells us **how strong the evidence is against H‚ÇÄ**

## 2Ô∏è‚É£ Meaning of P-Value
The p-value measures:
- How likely the observed sample result is
- When the null hypothesis is assumed to be true

A **small p-value** means:
- The observed result is **very unlikely** under H‚ÇÄ
- There is **strong evidence against H‚ÇÄ**

A **large p-value** means:
- The observed result is **likely** under H‚ÇÄ
- There is **weak evidence against H‚ÇÄ**


## 3Ô∏è‚É£ Intuition Behind P-Value
Think of p-value as:
- A **measure of surprise**

If the p-value is very small:
- The result is too surprising to occur by chance
- The null hypothesis may not be true

If the p-value is large:
- The result is not surprising
- The null hypothesis can still be true

## 4Ô∏è‚É£ Example to Understand P-Value

### Coin Toss Example
Assume:
H‚ÇÄ : Coin is fair

If we toss a coin many times and get mostly heads:
- The p-value tells us **how likely such a result is**
- If the p-value is very small, the coin is probably **not fair**

## 5Ô∏è‚É£ Decision Rule Using P-Value

We compare the p-value with a **significance level (Œ±)**.

If p-value ‚â§ Œ± ‚Üí Reject H‚ÇÄ
If p-value > Œ± ‚Üí Do not reject H‚ÇÄ


## 6Ô∏è‚É£ Common Significance Levels

Frequently used values of Œ±:
Œ± = 0.05 (5% level)
Œ± = 0.01 (1% level)

## 7Ô∏è‚É£ Interpretation of P-Value

- p-value = 0.03  
  ‚Üí There is a **3% chance** of observing such data if H‚ÇÄ is true  

- p-value = 0.40  
  ‚Üí There is a **40% chance** of observing such data if H‚ÇÄ is true  

Smaller p-values indicate **stronger evidence** against H‚ÇÄ.


## 8Ô∏è‚É£ Important Points about P-Value
- P-value is **not the probability that H‚ÇÄ is true**
- It only measures evidence **against H‚ÇÄ**
- Smaller p-value ‚Üí stronger evidence
- Used directly in **hypothesis testing decisions**
