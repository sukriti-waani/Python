## 2Ô∏è‚É£ Confidence Interval
A **confidence interval (CI)** is a **range of values** used to estimate a population parameter.

It provides an interval within which the **true population parameter is likely to lie**.

### Interpretation
If we say a **95% confidence interval**, it means:
- We are **95% confident** that the interval contains the true population parameter

It does **not** mean:
- There is a 95% probability that the parameter lies in the interval
