## Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with jusr a few lined of code.

```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn

tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title('Total Bill vs Tips')
plt.show()
```

```python
# Line Plot

sns.lineplot(x = 'size', y = 'total_bill', data = tips)
plt.title('Line Plot of Total Bill by Size')
plt.show()
```

```python
# categorical Plots

# Bar Plot
sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
# Box Plot

sns.boxplot(x = "day", y = "total_bill", data = tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
# Violin Plot

sns.violinplot(x = "day", y = "total_bill", data = tips)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
# Histograms

sns.histplot(tips['total_bill'], bins=10, kde=True)
# bins=10 â†’ divides the data range into 10 bars (intervals).
# kde=True â†’ draws a smooth density curve over the histogram.
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# KDE Plot

sns.kdeplot(tips['total_bill'], fill=True)
# Plots a KDE (Kernel Density Estimate) curve for 'total_bill'.
# KDE shows the smooth shape of the data distribution.
# fill=True â†’ fills the area under the curve with color.

plt.title('KDE Plot of Total Bill')
plt.show()
```

```python
# Pair Plot

sns.pairplot(tips)
```

```python
# Heatmap

corr = tips[['total_bill', 'tip', 'size']].corr()
# Selects the columns: 'total_bill', 'tip', and 'size'
# .corr() calculates the correlation between these columns
# The result is a correlation matrix showing how strongly the values are related

corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
# annot=True â†’ displays the correlation values on the heatmap
# cmap='coolwarm' â†’ sets the color scheme for the heatmap
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
import pandas as pd

sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# Plot total sales by product

plt.figure(figsize=(10,6))
# Creates a figure (canvas) of size 10x6 inches.

sns.barplot(x='Product', y='Total_Sales', data=sales_df, estimator=sum)
# Draws a bar plot.
# x = 'Product' â†’ product names on x-axis.
# y = 'Total_Sales' â†’ height of each bar.
# data = sales_df â†’ uses the sales_df DataFrame.
# estimator = sum â†’ adds sales values for each product before plotting.

plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```
