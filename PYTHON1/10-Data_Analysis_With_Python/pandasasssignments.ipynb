```python
import numpy as np
import pandas as pd
```

### 1: DataFrame Creation and Indexing

```python
# 4 columns, 6 rows, set index to first column

df = pd.DataFrame(np.random.randint(1, 100, size=(6, 4)), columns=['A', 'B', 'C', 'D'])
print("Original DataFrame:")
print(df)

# Set the index to be the first column
df.set_index('A', inplace=True)
print("DataFrame with new index:", df)
```

```python
# DataFrame with custom index and access element

# DataFrame with given columns and index
df2 = pd.DataFrame(
    np.random.randint(1, 100, size=(3, 3)),
    columns=['A', 'B', 'C'],
    index=['X', 'Y', 'Z']
)

print("DataFrame:\n", df2)

# Access element at row 'Y' and column 'B'
value = df2.loc['Y', 'B']
print("\nElement at row 'Y', column 'B':", value)
```

### 2: DataFrame Operations

```python
# Add new column as product of first two

df3 = pd.DataFrame(
    np.random.randint(1, 10, size=(5, 3)),
    columns=['col1', 'col2', 'col3']
)
print("Original DataFrame:\n", df3)

# New column = product of col1 and col2
df3['product'] = df3['col1'] * df3['col2']
print("\nAfter adding 'product' column:\n", df3)
```

```python
# Row-wise and column-wise sum

df4 = pd.DataFrame(
    np.random.randint(1, 10, size=(4, 3)),
    columns=['A', 'B', 'C']
)
print("DataFrame:\n", df4)

# Row-wise sum (axis=1)
row_sum = df4.sum(axis=1)
print("\nRow-wise sum:\n", row_sum)

# Column-wise sum (axis=0 â€“ default)
col_sum = df4.sum(axis=0)
print("\nColumn-wise sum:\n", col_sum)
```

### 3: Data Cleaning

```python
# Fill NaN with column mean

df5 = pd.DataFrame(
    np.random.randint(1, 20, size=(5, 3)).astype(float),
    columns=['A', 'B', 'C']
)

# Introduce some NaN values
df5.iloc[1, 0] = np.nan
df5.iloc[3, 2] = np.nan

print("DataFrame with NaN:\n", df5)

# Fill NaN with column mean
df5_filled = df5.fillna(df5.mean())
print("\nAfter filling NaN with column mean:\n", df5_filled)
```

```python
# Drop rows with any NaN

df6 = pd.DataFrame(
    np.random.randint(1, 20, size=(6, 4)).astype(float),
    columns=['A', 'B', 'C', 'D']
)

# Introduce NaN values
df6.iloc[0, 1] = np.nan
df6.iloc[4, 3] = np.nan

print("DataFrame with NaN:\n", df6)

# Drop rows having any NaN
df6_dropped = df6.dropna()
print("\nAfter dropping rows with NaN:\n", df6_dropped)
```

### 4: Data Aggregation

```python
# Group by 'Category' and aggregate

categories = np.random.choice(['A', 'B', 'C'], size=10)
values = np.random.randint(1, 50, size=10)

df7 = pd.DataFrame({
    'Category': categories,
    'Value': values
})
print("Original DataFrame:\n", df7)

# Group by 'Category' and compute sum and mean
grouped = df7.groupby('Category')['Value'].agg(['sum', 'mean'])
print("\nSum and mean of 'Value' by Category:\n", grouped)
```

```python
# Group by 'Category' and total sales

products = ['P1', 'P2', 'P3', 'P4', 'P5']
categories = np.random.choice(['Electronics', 'Clothes', 'Groceries'], size=5)
sales = np.random.randint(100, 1000, size=5)

df8 = pd.DataFrame({
    'Product': products,
    'Category': categories,
    'Sales': sales
})
print("Sales Data:\n", df8)

# Group by 'Category' and sum Sales
total_sales = df8.groupby('Category')['Sales'].sum()
print("\nTotal sales by Category:\n", total_sales)
```

### 5: Merging DataFrames

```python
# Merge on common column

df_left = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['A', 'B', 'C']
})

df_right = pd.DataFrame({
    'ID': [1, 2, 4],
    'Score': [90, 85, 75]
})

print("Left:\n", df_left)
print("\nRight:\n", df_right)

merged = pd.merge(df_left, df_right, on='ID', how='inner')
print("\nMerged on 'ID':\n", merged)
```

```python
# Concatenate along rows and columns

df_a = pd.DataFrame(np.random.randint(1, 10, size=(3, 2)), columns = ['X', 'Y'])

df_b = pd.DataFrame(np.random.randint(1, 10, size=(3, 2)), columns = ['X', 'Y'])


print("df_a:\n", df_a)
print("\ndf_b:\n", df_b)

# Concatenate along rows
concatRows = pd.concat([df_a, df_b], axis = 0)
print("Concatenated along rows: \n", concatRows)

# Concatenate along columns
concatCols = pd.concat([df_a, df_b], axis = 1)
print("Concatenated along columns: \n", concatCols)
```
