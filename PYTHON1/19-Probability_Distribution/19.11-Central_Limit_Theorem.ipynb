The **Central Limit Theorem (CLT)** is a fundamental theorem in statistics that explains the behavior of the **sampling distribution of the sample mean**.

It states that when we take a **large number of samples** from a population and compute their means, the distribution of those sample means tends to follow a **normal (Gaussian) distribution**.

This happens **regardless of the original population distribution**, provided the sample size is sufficiently large.

## 2Ô∏è‚É£ Sampling Distribution
The **sampling distribution** is the probability distribution of a statistic (such as the mean) calculated from **all possible samples of a given size** drawn from a population.

In CLT, we mainly focus on the **sampling distribution of the mean**.
