In statistics, an **estimate** is a **specified observed numerical value** used to estimate an **unknown population parameter**.

Since population parameters (like population mean Œº) are usually unknown, we use **sample statistics** to estimate them.

- **Population parameter**: A numerical value that describes the population

## 5Ô∏è‚É£ Limitations of Point Estimation
- It does not show **how accurate** the estimate is  
- It does not indicate **possible error or uncertainty**  
- The true population parameter may not be exactly equal to the point estimate  

Because of these limitations, **interval estimation** is used.

## 6Ô∏è‚É£ Interval Estimation
An **interval estimate** is a **range of values** used to estimate an unknown population parameter.

Instead of giving a single value, it gives a **range within which the parameter is expected to lie**.

- The interval gives a **measure of uncertainty**
- A wider interval means **more uncertainty**
- A narrower interval means **more precision**
