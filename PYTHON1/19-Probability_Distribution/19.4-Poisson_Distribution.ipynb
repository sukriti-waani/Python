In probability theory and statistics, the **Poisson distribution** is a **discrete probability distribution** that expresses the probability of a **given number of events occurring in a fixed interval of time or space**, provided that:

- Events occur **independently**
- Events occur at a **constant average rate**
- Two events cannot occur at exactly the same instant
