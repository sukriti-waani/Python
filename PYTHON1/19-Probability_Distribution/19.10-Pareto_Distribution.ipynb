## 1Ô∏è‚É£ Introduction to Pareto Distribution

The **Pareto distribution**, named after the Italian economist **Vilfredo Pareto**, is a **power-law probability distribution**.

It is commonly used to describe situations where:
- A **small number of causes** produce a **large portion of the effect**
- Most values are small, and a few values are extremely large

This type of distribution is also referred to as a **non-Gaussian distribution**.

## 2Ô∏è‚É£ Nature of Pareto (Power Law) Distribution
- It is a **continuous distribution**
- It is **highly right-skewed**
- It exhibits **long-tail behavior**
- Mean and variance can be very large or sometimes undefined
- The distribution is not symmetric

The long tail is clearly shown in the graph on page 1 of the notes.

## 3Ô∏è‚É£ Power Law and Long-Tail Behavior
In a power-law distribution:
- The **head** contains a few very large values
- The **tail** contains many small values extending far to the right

This is known as a **long-tail distribution**, where rare but extreme values have a strong impact.

## 4Ô∏è‚É£ 80‚Äì20 Rule (Pareto Principle)
The Pareto distribution follows the **80‚Äì20 rule**, also known as the **Pareto principle**.

It states that:
- **80% of outcomes come from 20% of causes**

## 5Ô∏è‚É£ Mathematical Representation
The probability of observing a value `x` in a power-law distribution is given by:

P(X = x) ‚àù x^(‚àíŒ±)

Where:
- `Œ± > 0` is the shape parameter
- Larger values of `x` have much smaller probability

## 6Ô∏è‚É£ Pareto Type I Distribution
- Different curves correspond to different values of `Œ±`
- As `Œ±` increases, the curve drops faster
- Smaller `Œ±` values produce heavier tails

This explains why some systems have extreme imbalance.

## 7Ô∏è‚É£ Transformation to Log-Normal Distribution
Y = ln(X)

- Reduces skewness
- Makes the distribution more symmetric
- Converts power-law behavior closer to a **normal (Gaussian) distribution**

This transformation is shown visually as:
Power Law ‚Üí log transform ‚Üí Log-Normal / Normal-like curve

## 8Ô∏è‚É£ Box-Cox Transformation
Purpose:
- Reduce right skewness
- Stabilize variance
- Make data more suitable for modeling

It is a generalization of the log transformation and is useful in machine learning.

## 9Ô∏è‚É£ QQ Plot Interpretation
After transformation:
- A **QQ plot** is used to check normality
- If points lie close to a straight line, the transformed data is approximately normal

This confirms whether the transformation was effective.
