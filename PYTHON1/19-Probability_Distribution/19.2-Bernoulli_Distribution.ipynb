## 1Ô∏è‚É£ Introduction to Bernoulli Distribution
The **Bernoulli distribution** is the **simplest discrete probability distribution**.

It represents the probability distribution of a random variable that has **exactly two possible outcomes**:
- **Success** with probability `p`
- **Failure** with probability `q = 1 ‚àí p`

It is commonly used to model **binary outcomes**, such as:
- Coin toss (Head / Tail)
- Pass / Fail
- Yes / No
- Success / Failure

## 2Ô∏è‚É£ Bernoulli Random Variable
A Bernoulli random variable `X` takes only **two values**:

X ‚àà {0, 1}

- `X = 1` ‚Üí Success  
- `X = 0` ‚Üí Failure  


## 3Ô∏è‚É£ Parameters of Bernoulli Distribution

The Bernoulli distribution has **one parameter**:

0 ‚â§ p ‚â§ 1

Where:
- `p = P(X = 1)` ‚Üí probability of success
- `q = P(X = 0) = 1 ‚àí p` ‚Üí probability of failure


## 4Ô∏è‚É£ Probability Mass Function (PMF)

P(X = k) = p^k (1 ‚àí p)^(1 ‚àí k)

Where:
k ‚àà {0, 1}


#### Simplified PMF

P(X = 1) = p
P(X = 0) = 1 ‚àí p = q

## 7Ô∏è‚É£ Median of Bernoulli Distribution

The **median** depends on the value of `p`:

Median = 0 if p < 0.5
Median = 0.5 if p = 0.5
Median = 1 if p > 0.5

## 8Ô∏è‚É£ Mode of Bernoulli Distribution
The **mode** is the value with the **highest probability**.

Mode = 1 if p > q
Mode = 0 if q > p
