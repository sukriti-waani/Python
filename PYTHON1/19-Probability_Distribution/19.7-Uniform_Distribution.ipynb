The **uniform distribution** is a probability distribution in which **all outcomes are equally likely** within a given range.

There are **two types of uniform distributions**:
1. **Continuous Uniform Distribution**
2. **Discrete Uniform Distribution**

## 2Ô∏è‚É£ Continuous Uniform Distribution
A **continuous uniform distribution** is a probability distribution in which a continuous random variable can take **any value between two given bounds**, and **all values in this interval are equally likely**.

It is also called a **rectangular distribution**.


## 3Ô∏è‚É£ Nature of Continuous Uniform Distribution

- It is a **continuous random variable**

## 1Ô∏è‚É£1Ô∏è‚É£ Discrete Uniform Distribution
A **discrete uniform distribution** is a probability distribution in which a **finite number of discrete outcomes are equally likely**.

Each outcome has the **same probability**.

## 1Ô∏è‚É£2Ô∏è‚É£ Nature of Discrete Uniform Distribution

- It is a **discrete random variable**
- Total number of outcomes is **finite**
- Probability Mass Function (PMF) is used

## 1Ô∏è‚É£3Ô∏è‚É£ Example: Rolling a Fair Dice
Possible outcomes: {1, 2, 3, 4, 5, 6}

Number of outcomes: n = 6

Probability of each outcome:

P(X = x) = 1 / n = 1 / 6

So: P(1) = P(2) = P(3) = P(4) = P(5) = P(6) = 1/6

## 1Ô∏è‚É£4Ô∏è‚É£ Parameters of Discrete Uniform Distribution
Let: X ~ U(a, b)

Where:
- `a` = minimum value
- `b` = maximum value
- `b ‚â• a`
- Total outcomes: n = b ‚àí a + 1

## 1Ô∏è‚É£5Ô∏è‚É£ Probability Mass Function (PMF)
P(X = x) = 1 / n

Where: x ‚àà {a, a+1, ..., b}

## 1Ô∏è‚É£6Ô∏è‚É£ Mean of Discrete Uniform Distribution
Mean = (a + b) / 2

## 1Ô∏è‚É£7Ô∏è‚É£ Median of Discrete Uniform Distribution
Median = (a + b) / 2

## 1Ô∏è‚É£8Ô∏è‚É£ Graphical Interpretation (Discrete Case)
- PMF graph consists of **equal-height bars**

## 1Ô∏è‚É£9Ô∏è‚É£ Important Properties

- All outcomes are **equally likely**
- Continuous case uses **PDF**
- Discrete case uses **PMF**
- Mean and median are the same
- Distribution is **symmetric**
