## 1Ô∏è‚É£ Introduction to Power Law Distribution
In statistics, a **power law distribution** describes a relationship between two quantities where a **relative change in one quantity results in a proportional relative change in another quantity**, independent of the initial size.

In simple words:
- A **small number of items contribute to a very large effect**
- Most values are small, and a few values are extremely large

## 2Ô∏è‚É£ Nature of Power Law Distribution
- It is a **continuous distribution**
- The distribution is **highly right-skewed**
- It has a **long tail**
- A few values dominate the entire distribution
- Mean and variance may be very large or undefined in some cases

## 3Ô∏è‚É£ Long-Tail Behavior
In a power law distribution:
- The **head** represents a few large values
- The **tail** represents many small values

This is called a **long-tail distribution**
## 4Ô∏è‚É£ 80‚Äì20 Rule (Pareto Principle)

Power law distributions often follow the **80‚Äì20 rule**, also called the **Pareto principle**.

This rule states:
- **80% of the effects come from 20% of the causes**

## 5Ô∏è‚É£ Real-World Examples of Power Law Distribution
Examples listed in the notes include:

1. **IPL Matches**
   - 20% of teams win 80% of the matches

2. **Wealth Distribution**
   - 80% of wealth is owned by 20% of the population

3. **Oil Resources**
   - 80% of total oil belongs to 20% of nations

4. **Language**
   - Frequency of words in most languages follows a power law

5. **Software Defects**
   - 20% of major defects cause 80% of system failures

These examples show how **few contributors dominate outcomes**.

## 7Ô∏è‚É£ Relationship with Log Transformation
When we apply **log transformation**:
Y = ln(X)

- The skewed power law curve becomes **more symmetric**
- The transformed data may resemble a **normal distribution**

## 8Ô∏è‚É£ Box-Cox Transformation
The notes also show **Box-Cox transformation** as a generalization of log transform.

Purpose:
- Reduce skewness
- Stabilize variance
- Make data closer to normal distribution

This helps in statistical modeling and machine learning.


## 9Ô∏è‚É£ QQ Plot Interpretation
After transformation:
- A **QQ plot** is used to compare data with a normal distribution
- If points lie close to a straight line, the transformed data is approximately normal

This confirms that power law data becomes manageable after transformation.
