## 1Ô∏è‚É£ Introduction to Binomial Distribution
In probability theory and statistics, the **binomial distribution** is a **discrete probability distribution** that models the **number of successes** in a fixed number of independent trials.

Each trial:
- Has only **two possible outcomes** (success or failure)
- Has the **same probability of success**

A single success/failure experiment is called a **Bernoulli trial**, and a sequence of such trials forms a **Bernoulli process**.

## 2Ô∏è‚É£ Conditions for Binomial Distribution

A random experiment follows a **binomial distribution** if:

1. The experiment consists of **n fixed trials**
2. Each trial has only **two outcomes** (success/failure)
3. Trials are **independent**
4. Probability of success **p remains constant**
5. Each trial is a **Bernoulli trial**

## 3Ô∏è‚É£ Binomial Random Variable

Let:
- `X` = number of successes in `n` trials  

Then:
X = 0, 1, 2, ..., n

## 4Ô∏è‚É£ Notation of Binomial Distribution

The binomial distribution is denoted as:
X ~ B(n, p)

Where:
- `n` = number of trials  
- `p` = probability of success in each trial  
- `q = 1 ‚àí p` = probability of failure  

## 5Ô∏è‚É£ Parameters of Binomial Distribution

- Number of trials:  
n ‚àà {0, 1, 2, 3, ..., n}

- Probability of success:
0 ‚â§ p ‚â§ 1

- Probability of failure:
q = 1 ‚àí p

## 6Ô∏è‚É£ Support of Binomial Distribution

The support (possible values of X) is:
k ‚àà {0, 1, 2, ..., n}

Where `k` represents the **number of successes**.

## 7Ô∏è‚É£ Probability Mass Function (PMF)
P(X = k) = C(n, k) ¬∑ p^k ¬∑ (1 ‚àí p)^(n ‚àí k)

Where:
C(n, k) = n! / [k!(n ‚àí k)!]

This is called the **binomial coefficient**.

## Example 1: Coin Toss Experiment
A coin is tossed **5 times**.  
Find the probability of getting **exactly 3 heads**.

Given: 
n = 5, p = 0.5, k = 3

P(X = 3) = C(5, 3) ¬∑ (0.5)^3 ¬∑ (0.5)^2 
= 10 √ó 0.125 √ó 0.25 
= 0.3125

## Example 2: Quality Control (from notes)

Find the probability of getting **exactly 2 defective items**.

Given: n = 10, p = 0.1, k = 2
